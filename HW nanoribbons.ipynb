# 1. Zigzag nanoribbon (8 atoms case)

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

###############
##b c   f g####
#a   d e   h###
##b c   f g####
#a   d e   h###
###############
V=0
d=np.sqrt(3)  # lattice distance in y direction is sqrt(3) (Setting 1 as atom-to-atom distance)
t=-1

R1=[(1/2),np.sqrt(3)/2]
R2=[1,0]
R3=[1/2,-np.sqrt(3)/2]

k=np.linspace(-np.pi/(d),np.pi/(d),100)
energyband=[]

def f(R,y):
    return np.exp(1j*np.dot([0,y],[R[0],R[1]]))   

M = np.zeros((8,8))
for i in range(len(k)):
    M[0,1] = t*(f(R1,k[i])+f(R3,k[i]))  
    M[1,0] = np.conjugate(M[0,1]) 
    M[1,2] = t*(f(R2,k[i]))  
    M[2,1] = np.conjugate(M[1,2])
    M[2,3] = t*(f(R1,k[i])+f(R3,k[i]))  
    M[3,2] = np.conjugate(M[2,3]) 
    M[3,4] = t*(f(R2,k[i]))  
    M[4,3] = np.conjugate(M[3,4])  
    M[4,5] = t*(f(R1,k[i])+f(R3,k[i]))  
    M[5,4] = np.conjugate(M[4,5])  
    M[5,6] = t*(f(R2,k[i]))  
    M[6,5] = np.conjugate(M[5,6])  
    M[6,7] = t*(f(R1,k[i])+f(R3,k[i]))  
    M[7,6] = np.conjugate(M[6,7])  
      
    w,v=LA.eig(M)
    w.sort()
    energyband.append(w)
             

```

```python
#print(M)
plt.figure()
plt.plot(k, energyband, label='energy band')
```
